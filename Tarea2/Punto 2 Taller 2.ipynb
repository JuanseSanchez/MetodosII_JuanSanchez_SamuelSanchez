```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
h = 0.01
t = np.arange(0,10+h,h)
```

```python
def function(x,t,k=1.0,v0=1.0,m=1.0):
    h = t[1]-t[0]
    a = -k*x[0]/ m
    x[1] = x[0] + v0*h+0.5*a*h**2
    for i in range(1,len(t)- 1):
        a = -k * x[i] / m
        x[i+1] = 2*x[i]-x[i-1]+a*h**2
    return x
```

```python
xt = np.zeros_like(t)
x = function(xt,t)
```

```python
plt.plot(t,x, label='Position')
plt.xlabel('Time')
plt.ylabel('Position')
plt.legend()
plt.show()
```
