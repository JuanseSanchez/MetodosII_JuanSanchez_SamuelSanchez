```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f(x,y):
    return x*(y**2)-2*(x**-1)*y-(x**-3)
```

```python
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
N = 100
x = np.linspace(np.sqrt(2),11,N)
h = x[1]-x[0]
```

```python
y2 = np.zeros_like(x)
x[0] = np.sqrt(2)
```

```python
GetEuler2(x,y,h)
```

```python
def dydt(y, x):
    return x**4 * y**2 - 2 * x**2 * y - 1

y_exact = odeint(dydt, y2[0], x).flatten()

# Graficar las soluciones
plt.plot(x, y2, label='Runge Kutta 2')
plt.plot(x, y_exact, label='Solución exacta (odeint)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Runge Kutta 2 y Solución Exacta')
plt.legend()
plt.show()
```

```python
def GetEuler3(x,y,h):
        for i in range(1,len(x)):
                k1 = f(x[i-1],y[i-1])
                k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
                k3 = f(x[i-1]+h,y[i-1] - h*k1 + 2*h*k2)
                y[i] = y[i-1] + (h/6)*(k1+4*k2+k3)
```

```python
y3 = np.zeros_like(x)
```

```python
GetEuler3(x,y3,h)
```

```python
plt.plot(x, y3, label='Runge Kutta 3')
plt.plot(x, y_exact, label='Solución exacta (odeint)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Runge Kutta 3 y Solución Exacta')
plt.legend()
plt.show()
```

```python
def GetEuler4(x,y,h):
        for i in range(1,len(x)):
                k1 = f(x[i-1],y[i-1])
                k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
                k3 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k2)
                k4 = f(x[i-1]+h,y[i-1] + h*k3)
                y[i] = y[i-1] + (h/6)*(k1+2*k2+2*k3+k4)
```

```python
y4 = np.zeros_like(x)
```

```python
GetEuler4(x,y4,h)
```

```python
plt.plot(x, y4, label='Runge Kutta 4')
plt.plot(x, y_exact, label='Solución exacta (odeint)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Runge Kutta 4 y Solución Exacta')
plt.legend()
plt.show()
```

```python
plt.plot(x,np.abs(y2-y_exact),label='Error Runge Kutta 2')
plt.plot(x,np.abs(y3-y_exact),label='Error Runge Kutta 3')
plt.plot(x,np.abs(y4-y_exact),label='Error Runge Kutta 4')
plt.legend()
```
