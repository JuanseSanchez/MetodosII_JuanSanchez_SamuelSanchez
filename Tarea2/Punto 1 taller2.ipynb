```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def fc(z):
    return 1j*z
```

```python
def ComplexEuler():
    global theta
    global h_
    global f

    theta +=h_
    f +=h_*fc(f)
```

```python
theta = 0
N = 100
Vueltas = 1

h_ = Vueltas *2*np.pi/N

f = 1

Solucion1 = np.array([theta,f])

for i in range(N):
    ComplexEuler()
    Solucion1 = np.vstack((Solucion1,np.array([theta,f])))
```

```python
Re = np.real(Solucion1[:,1])
Im = np.imag(Solucion1[:,1])

plt.plot(Re,Im)
```

```python
def ComplexEuler2():
    global theta
    global h_
    global f
    
    theta += h_
    
    k1 = fc(f)
    k2 = fc(f+h_*k1)
    
    f += 0.5*h_*(k1+k2)
```

```python
Solucion2 = np.array([theta,f])

for i in range(N):
    ComplexEuler2()
    Solucion2 = np.vstack((Solucion2,np.array([theta,f])))
```

```python
Re = np.real(Solucion1[:,1])
Im = np.imag(Solucion1[:,1])

plt.plot(Re,Im)

Re2 = np.real(Solucion2[:,1])
Im2 = np.imag(Solucion2[:,1])

plt.plot(Re2,Im2)
```

```python
def ComplexEuler3():
    global theta
    global h_
    global f

    theta += h_

    k1 = fc(f)
    k2 = fc(f+0.5*h_*k1)
    k3 = fc(f-h_*k1+2*h_*k2)

    f += (h_/6)*(k1+4*k2+k3)
```

```python
def ComplexEuler4():
    global theta
    global h_
    global f

    theta += h_

    k1 = fc(f)
    k2 = fc(f+0.5*h_*k1)
    k3 = fc(f+0.5*h_*k2)
    k4 = fc(f+h_*k3)

    f += (h_/6)*(k1+2*k2+2*k3+k4)
```

```python
Solucion3 = np.array([theta,f])

for i in range(N):
    ComplexEuler3()
    Solucion3 = np.vstack((Solucion3,np.array([theta,f])))
```

```python
Solucion4 = np.array([theta,f])

for i in range(N):
    ComplexEuler4()
    Solucion4 = np.vstack((Solucion4,np.array([theta,f])))
```

```python
Re = np.real(Solucion1[:,1])
Im = np.imag(Solucion1[:,1])

plt.plot(Re,Im)

Re2 = np.real(Solucion2[:,1])
Im2 = np.imag(Solucion2[:,1])

plt.plot(Re2,Im2)

Re3 = np.real(Solucion3[:,1])
Im3 = np.imag(Solucion3[:,1])

plt.plot(Re3,Im3)

Re4 = np.real(Solucion4[:,1])
Im4 = np.imag(Solucion4[:,1])

plt.plot(Re4,Im4)
```

```python
plt.plot(np.abs(Re2-Re),label='Runge-Kutta 2 Real')
plt.plot(np.abs(Im2-Im),label='Runge-Kutta 2 Imag')
plt.legend()
```

```python
plt.plot(np.abs(Re3-Re2),label='Runge-Kutta 3 Real')
plt.plot(np.abs(Im3-Im2),label='Runge-Kutta 3 Imag')
plt.legend()
```

```python
plt.plot(np.abs(Re4-Re3),label='Runge-Kutta 4 Real')
plt.plot(np.abs(Im4-Im3),label='Runge-Kutta 4 Imag')
plt.legend()
```

Podemos ver que el algoritmo de Runge Kutta cada vez es ms estable al aumentar su orden. Esta vez lo comprobamos con el tercer y cuarto orden.
