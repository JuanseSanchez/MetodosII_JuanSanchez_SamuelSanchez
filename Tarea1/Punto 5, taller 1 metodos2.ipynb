```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -1
dt1 = 1.1
dt2 = 1.5
dt3 = 1.9
```

```python
def FunExacta(a,t):
    return np.exp(a*t)
```

```python
def f(k,u0,deltat):
    return ((1+a*deltat)**k) *u0
```

```python
t_values = np.linspace(-10, 10, 100) 
plt.plot(t_values, FunExacta(a,t_values), label=f'e^at')
plt.title('Plot of the function f(t) = e^{at}')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.show()
```

```python
y1 = np.zeros_like(t_values)
y1[0] = 1
y2 = np.zeros_like(t_values)
y2[0] = 1
y3 = np.zeros_like(t_values)
y3[0] = 1
```

```python
for i in range(1, len(t_values)):
    y1[i] = f(i, y1[i-1], dt1)
    y2[i] = f(i, y2[i-1], dt2)
    y3[i] = f(i, y3[i-1], dt3)
```

```python
plt.plot(t_values,y1)
```

```python
plt.plot(t_values,y2)
```

```python
plt.plot(t_values,y3)
```
